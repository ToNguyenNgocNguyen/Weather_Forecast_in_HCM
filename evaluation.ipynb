```python
import os
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("seaborn-v0_8-paper") 
```

```python
# Directory containing the files
directory = 'train_logs'

# Read file logs.csv
df = pd.read_csv(os.path.join(directory, "logs.csv"))
```

```python
for col in df.columns[1:]:
    plt.figure(figsize=(10, 4))
    plt.plot(df['Epoch'], df[col])
    # plt.title(f'{col} change over epochs')
    plt.title(' ')
    plt.ylabel(col)
    plt.xlabel('epochs')
    plt.savefig(f"evaluation_plot/{col}_evaluation")
```
